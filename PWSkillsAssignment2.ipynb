#### Q3. How do you convert one data type to another in Python?

```python
# we can convert the one datatype to another datatype using the technique called "Type Casting"

# for example if you have a string 'a' as '100' , if you want to convert the string '100' to the integer then we use 
# typecasting methods

a='100'
print("Type of variable a",type(a)) # In the output we got the type of variable 'a' is string

# Now, we need to use typecasting methods to convert the string into integer

b=int(a) # typecasted the string into integer using the method int() and stored the result in the variable 'b'

print("Type of variable b",type(b)) # In the output the type of variable 'b' is shown as "int"

# Note:
#     If the string contain any alphabets while typecasting then we will get an error as we can't convert alphabets into integers
```

#### Q6. What is a complex number in mathematics, and how is it represented in Python?

```python
# Complex number is the combination of real and imaginary number 

# ex. 3+2j --> 3 is the real number and 2 is the imaginary number 

# In python we write the complex number as below:
    
a=2+3j
print(type(a))
print("The real part is:",a.real)
print("The imaginary part is:",a.imag)
```
